```python
from sklearn.preprocessing import StandardScaler

features = df_new.drop(columns=['Global_Avg'])
target = df_new['Global_Avg']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

# Anamaly Detection

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.01)
y_pred_outliers = lof.fit_predict(X_scaled)
df_new['LOF_Outlier'] = y_pred_outliers
df_new['LOF_Label'] = df_new['LOF_Outlier'].map({1: 'Normal', -1: 'Outlier'})
features_to_plot = df_new.drop(columns=['LOF_Outlier', 'LOF_Label']).columns

plt.figure(figsize=(18, 12))
for i, col in enumerate(features_to_plot):
    plt.subplot(4, 4, i + 1)
    sns.boxplot(data=df_new, x='LOF_Label', y=col)
    plt.title(col)
    plt.tight_layout()
plt.suptitle("Box Plots of Features: Normal vs. LOF Outliers", y=1.02, fontsize=16)
plt.show()
```

```python
df_new['LOF_Outlier'] = y_pred_outliers  # -1 = outlier, 1 = inlier
df_cleaned = df_new[df_new['LOF_Outlier'] == 1].copy()

df_cleaned.drop(columns=['LOF_Outlier', 'LOF_Label'], inplace=True, errors='ignore')
plt.suptitle("Box Plots of Features: Normal vs. LOF Outliers", y=1.02, fontsize=16)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred_outliers, cmap='coolwarm', s=2)
plt.title("Local Outlier Factor (LOF) Detection")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

# Construction of ML models

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred_lin = lin_reg.predict(X_test)

print('Linear Regression Performance\n')
print(f"R Score: {r2_score(y_test, y_pred_lin):.4f}")
print(f"MAE     : {mean_absolute_error(y_test, y_pred_lin):.2f}")
print(f"RMSE    : {np.sqrt(mean_squared_error(y_test, y_pred_lin)):.2f}")

plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred_lin, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression: Actual vs Predicted')
plt.show()
```

Decision Tree Regressor

```python
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train, y_train)
y_pred_tree = tree_reg.predict(X_test)

print('Decision Tree Performance\n')
print(f"R Score: {r2_score(y_test, y_pred_tree):.4f}")
print(f"MAE     : {mean_absolute_error(y_test, y_pred_tree):.2f}")
print(f"RMSE    : {np.sqrt(mean_squared_error(y_test, y_pred_tree)):.2f}")

plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred_tree, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Decision Tree: Actual vs Predicted')
plt.show()
```

Random Forest Regressor

```python
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(X_train, y_train)
y_pred_forest = forest_reg.predict(X_test)

print('Random Forest Performance\n')
print(f"R Score: {r2_score(y_test, y_pred_forest):.4f}")
print(f"MAE     : {mean_absolute_error(y_test, y_pred_forest):.2f}")
print(f"RMSE    : {np.sqrt(mean_squared_error(y_test, y_pred_forest)):.2f}")

plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred_forest, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Random Forest: Actual vs Predicted')
plt.show()
```

Gradient Boosting Regressor

```python
gb_reg = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_reg.fit(X_train, y_train)
y_pred_gb = gb_reg.predict(X_test)

print('Gradient Boosting Performance\n')
print(f"R Score: {r2_score(y_test, y_pred_gb):.4f}")
print(f"MAE     : {mean_absolute_error(y_test, y_pred_gb):.2f}")
print(f"RMSE    : {np.sqrt(mean_squared_error(y_test, y_pred_gb)):.2f}")

plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred_gb, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Gradient Boosting: Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import (
    r2_score, mean_absolute_error, mean_squared_error,
    median_absolute_error, max_error, explained_variance_score
)

def evaluate_model_extended(model, X_test, y_test, name="Model"):
    y_pred = model.predict(X_test)

    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    med_ae = median_absolute_error(y_test, y_pred)
    max_err = max_error(y_test, y_pred)
    evs = explained_variance_score(y_test, y_pred)

    print(f" {name} Evaluation")
    print(f"R Score              : {r2:.4f}")
    print(f"Explained Variance    : {evs:.4f}")
    print(f"MAE (Mean Abs Error)  : {mae:.2f}")
    print(f"MSE (Mean Sq Error)   : {mse:.2f}")
    print(f"RMSE (Root MSE)       : {rmse:.2f}")
    print(f"Median Abs Error      : {med_ae:.2f}")
    print(f"Max Error             : {max_err:.2f}")
    print("-" * 50)

    plt.figure(figsize=(6, 4))
    plt.scatter(y_test, y_pred, alpha=0.3, color="teal")
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"{name}: Actual vs Predicted")
    plt.tight_layout()
    plt.show()

evaluate_model_extended(lin_reg, X_test, y_test, name="Linear Regression")
evaluate_model_extended(tree_reg, X_test, y_test, name="Decision Tree Regressor")
evaluate_model_extended(forest_reg, X_test, y_test, name="Random Forest Regressor")
evaluate_model_extended(gb_reg, X_test, y_test, name="Gradient Boosting Regressor")
evaluate_model_extended(knn_reg, X_test, y_test, name="K-Nearest Neighbors Regressor")
```

# HyperparameterTuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint
rf = RandomForestRegressor(random_state=42)

param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,
    scoring='r2',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
best_rf_model = random_search.best_estimator_

print("Best Parameters:", random_search.best_params_)
print(f"Best Cross-Validation R Score: {random_search.best_score_:.4f}")
```

# Prediction & Final result

```python
y_pred_final = best_rf_model.predict(X_test)

r2_final = r2_score(y_test, y_pred_final)
mae_final = mean_absolute_error(y_test, y_pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))

print("\n Final Model Test Set Performance")
print(f"R Score: {r2_final:.4f}")
print(f"MAE     : {mae_final:.2f}")
print(f"RMSE    : {rmse_final:.2f}")

plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred_final, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Final Model: Actual vs Predicted')
plt.grid(True)
plt.show()
```
